```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
import glob
# Load all 40 raw CSVs from your raw folder
files = glob.glob("C:/23321A0533/bluestock/data/raw/02_nav_history.csv")

df_list = []
for f in files:
    temp = pd.read_csv(f)
    temp['scheme_name'] = f.split("/")[-1].replace(".csv","")  # add scheme name from filename
    df_list.append(temp)

df = pd.concat(df_list, ignore_index=True)
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['scheme_name','date'])
```

```python
# Remove extreme outliers in daily returns
# Keep only values between -20% and +20% (typical range for funds)
df['daily_return'] = df['daily_return'].clip(lower=-0.20, upper=0.20)

# Drop rows where NAV or daily_return is missing
df = df.dropna(subset=['nav','daily_return'])

# Validate again
plt.figure(figsize=(10,6))
df['daily_return'].hist(bins=100)
plt.title("Cleaned Distribution of Daily Returns (All Schemes)")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.show()

# Summary stats after cleaning
daily_stats_clean = df.groupby('scheme_name')['daily_return'].describe()
daily_stats_clean.head()
```

```python
# Sharpe Ratio calculation
rf = 0.065  # annual risk-free rate (RBI repo proxy)

sharpe_results = []

for scheme, group in df.groupby('scheme_name'):
    # Annualized mean return
    mean_return = group['daily_return'].mean() * 252
    
    # Annualized standard deviation
    std_return = group['daily_return'].std() * np.sqrt(252)
    
    # Sharpe Ratio
    sharpe = (mean_return - rf) / std_return if std_return else np.nan
    
    sharpe_results.append([scheme, sharpe])

# Build DataFrame
sharpe_df = pd.DataFrame(sharpe_results, columns=['scheme','Sharpe'])
sharpe_df.to_csv("sharpe.csv", index=False)
# Rank funds by Sharpe Ratio
sharpe_df = sharpe_df.sort_values('Sharpe', ascending=False)
sharpe_df.head(10)  # top 10 funds
```

```python
# Sortino Ratio calculation
def downside_std(returns):
    return returns[returns < 0].std()

sortino_results = []

for scheme, group in df.groupby('scheme_name'):
    # Annualized mean return
    mean_return = group['daily_return'].mean() * 252
    
    # Downside deviation (annualized)
    downside = downside_std(group['daily_return']) * np.sqrt(252)
    
    # Sortino Ratio
    sortino = (mean_return - rf) / downside if downside else np.nan
    
    sortino_results.append([scheme, sortino])

# Build DataFrame
sortino_df = pd.DataFrame(sortino_results, columns=['scheme','Sortino'])

# Rank funds by Sortino Ratio
sortino_df = sortino_df.sort_values('Sortino', ascending=False)
sortino_df.head(10)  # top 10 funds
```

```python
# -------------------------------
# Cell 8: Maximum Drawdown
# -------------------------------

import pandas as pd

mdd_results = []

for scheme, group in df.groupby('scheme_name'):
    group = group.sort_values('date').reset_index(drop=True)
    
    # Running maximum NAV
    running_max = group['nav'].cummax()
    
    # Drawdown = NAV / running_max - 1
    drawdown = group['nav'] / running_max - 1
    
    # Maximum drawdown value
    mdd = drawdown.min()
    
    # Date of worst drawdown
    worst_date = group.loc[drawdown.idxmin(), 'date']
    
    # Date when running max was last updated before worst drawdown
    peak_date = group.loc[:drawdown.idxmin(), 'nav'].idxmax()
    peak_date = group.loc[peak_date, 'date']
    
    mdd_results.append([scheme, mdd, peak_date, worst_date])

# Build DataFrame
mdd_df = pd.DataFrame(mdd_results, columns=['scheme','MaxDrawdown','PeakDate','WorstDate'])

# Save results
mdd_df.to_csv("max_drawdown.csv", index=False)

# Preview results
mdd_df.sort_values('MaxDrawdown').head(10)
```

```python
import pandas as pd
from scipy.stats import linregress

# -------------------------------
# 1. Load datasets
# -------------------------------
nav_history = pd.read_csv("02_nav_history.csv")   # columns: amfi_code, date, nav
benchmark_df = pd.read_csv("10_benchmark_indices.csv")  # columns: date, index_name, close_value

# Choose benchmark (example: NIFTY 100)
benchmark_df = benchmark_df[benchmark_df['index_name'] == "NIFTY100"].copy()

# Calculate daily benchmark returns
benchmark_df['bench_return'] = benchmark_df['close_value'].pct_change()
benchmark_df = benchmark_df[['date','bench_return']].dropna()

# -------------------------------
# 2. Loop through each scheme
# -------------------------------
results = []

for amfi, group in nav_history.groupby('amfi_code'):
    group = group.sort_values('date').reset_index(drop=True)

    # --- Sharpe ---
    daily_return = group['nav'].pct_change().dropna()
    mean_return = daily_return.mean() * 252
    vol = daily_return.std() * (252**0.5)
    sharpe = mean_return / vol if vol != 0 else None

    # --- Alpha/Beta ---
    merged = pd.merge(group, benchmark_df, on='date', how='inner')
    merged['fund_return'] = merged['nav'].pct_change()
    merged = merged.dropna(subset=['fund_return','bench_return'])
    if len(merged) > 50:
        slope, intercept, _, _, _ = linregress(merged['bench_return'], merged['fund_return'])
        alpha = intercept * 252
        beta = slope
    else:
        alpha, beta = None, None

    # --- Max Drawdown ---
    running_max = group['nav'].cummax()
    drawdown = group['nav'] / running_max - 1
    mdd = drawdown.min()
    peak_date = group.loc[:drawdown.idxmin(), 'nav'].idxmax()
    peak_date = group.loc[peak_date, 'date']
    worst_date = group.loc[drawdown.idxmin(), 'date']

    results.append([amfi, sharpe, alpha, beta, mdd, peak_date, worst_date])

# -------------------------------
# 3. Save outputs
# -------------------------------
df_out = pd.DataFrame(results, columns=[
    'amfi_code','Sharpe','Alpha','Beta','MaxDrawdown','PeakDate','WorstDate'
])
df_out.to_csv("metrics_all.csv", index=False)

print("Metrics saved to metrics_all.csv")
print(df_out.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load scorecard (with scheme_name, metrics, expense_ratio, return_3yr_pct, etc.)
score_df = pd.read_csv("fund_scorecard.csv")

# --- Add FundScore calculation ---
score_df['rank_return'] = score_df['return_3yr_pct'].rank(ascending=False)
score_df['rank_sharpe'] = score_df['Sharpe'].rank(ascending=False)
score_df['rank_alpha'] = score_df['Alpha'].rank(ascending=False)
score_df['rank_expense'] = score_df['expense_ratio_pct'].rank(ascending=True)
score_df['rank_mdd'] = score_df['MaxDrawdown'].rank(ascending=True)

for col in ['rank_return','rank_sharpe','rank_alpha','rank_expense','rank_mdd']:
    score_df[col] = 100 * (1 - (score_df[col]-1)/(len(score_df)-1))

score_df['FundScore'] = (
    0.30 * score_df['rank_return'] +
    0.25 * score_df['rank_sharpe'] +
    0.20 * score_df['rank_alpha'] +
    0.15 * score_df['rank_expense'] +
    0.10 * score_df['rank_mdd']
)

# --- Pick top 5 funds by FundScore ---
top5 = score_df.sort_values("FundScore", ascending=False).head(5)['amfi_code'].tolist()

# Load NAV history and benchmark data
nav_history = pd.read_csv("02_nav_history.csv")   # amfi_code, date, nav
benchmark_df = pd.read_csv("10_benchmark_indices.csv")  # date, index_name, close_value

# Filter last 3 years
cutoff = pd.to_datetime("2023-06-01")
nav_history['date'] = pd.to_datetime(nav_history['date'])
benchmark_df['date'] = pd.to_datetime(benchmark_df['date'])
nav_history = nav_history[nav_history['date'] >= cutoff]
benchmark_df = benchmark_df[benchmark_df['date'] >= cutoff]

# Prepare benchmark returns
benchmarks = {}
for idx in ["NIFTY50","NIFTY100"]:
    df = benchmark_df[benchmark_df['index_name'] == idx].copy()
    df['return'] = df['close_value'].pct_change()
    df = df[['date','return']].dropna()
    benchmarks[idx] = df

# --- Plot NAV comparison ---
plt.figure(figsize=(12,6))
for amfi in top5:
    fund = nav_history[nav_history['amfi_code'] == amfi].copy()
    fund['cum_return'] = (1 + fund['nav'].pct_change().fillna(0)).cumprod()
    plt.plot(fund['date'], fund['cum_return'], label=f"Fund {amfi}")

for idx in benchmarks:
    bm = benchmarks[idx].copy()
    bm['cum_return'] = (1 + bm['return']).cumprod()
    plt.plot(bm['date'], bm['cum_return'], label=idx, linewidth=2, linestyle='--')

plt.title("Top 5 Funds vs NIFTY 50 & NIFTY 100 (3 Years)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.show()

# --- Compute tracking error ---
tracking_errors = []
for amfi in top5:
    fund = nav_history[nav_history['amfi_code'] == amfi].copy()
    fund['fund_return'] = fund['nav'].pct_change()
    for idx in benchmarks:
        merged = pd.merge(fund[['date','fund_return']], benchmarks[idx], on='date', how='inner')
        te = (merged['fund_return'] - merged['return']).std() * (252**0.5)
        tracking_errors.append([amfi, idx, te])

tracking_df = pd.DataFrame(tracking_errors, columns=['amfi_code','benchmark','tracking_error'])
print(tracking_df)
```
